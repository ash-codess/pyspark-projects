```
from pyspark import SparkConf, SparkContext
from pyspark.sql.functions import col, lit
conf = SparkConf().setAppName("office-data")
sc = SparkContext.getOrCreate(conf=conf)
```

```
url = "https://raw.githubusercontent.com/ash-codess/pyspark-projects/master/data/office_data.csv"
from pyspark import SparkFiles
spark.sparkContext.addFile(url)

df = spark.read.csv("file://"+SparkFiles.get("office_data.csv"), header=True, inferSchema= True)
df.show()
```

```
#Total number of employees in the company
df.count()
```

```
#Total number and name of departments in the company
df1 = df.select("department").dropDuplicates(["department"]).count()
df2 = df.select("department").dropDuplicates(["department"]).show()
print("Number of departments: ", df1, df2)
```

```
#Total number of employees in each department
df.groupBy('department').count().show()
```

```
#Total number of employees in each state
df.groupBy('state').count().show()
```

```
#Total number of employees in each state and department
df.groupBy('state', 'department').count().show()
```

```
#Min and max salary in each department. Arrange in asc
from pyspark.sql.functions import min, max, avg
df.groupBy("department").agg(min("salary").alias("min"), max("salary").alias("max")).orderBy(col("max").asc(), col("min").asc()).show()
```

```
#Print names of employees who are working in NY state under finanance department whose bonsuses are grater than avergae bonuses of employees in NY
avgBonus = df.filter(df.state == "NY").groupBy("state").agg(avg("bonus").alias("avg_bonus")).select("avg_bonus").collect()[0]['avg_bonus']
df.filter((df.state == "NY") & (df.department == "Finance") & (df.bonus > avgBonus)).show()
```
